```python
# 9- Find the mean and standard deviation of height using the simple operations only. (np.sum) 
mean = heights.sum()/heights.shape[0]
print('Mean:', mean)
```

```python
diff = heights - mean
squared = np.power(diff, 2)
summ = np.sum(squared)
n = heights.shape[0]
s = np.sqrt(summ / (n))
print('Standard Deviation:',s)
```
